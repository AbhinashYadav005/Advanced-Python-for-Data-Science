```python
import sqlite3
#connect to the database
conn = sqlite3.connect("mydb.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
        CREATE TABLE IF NOT EXISTS students (
            id INTEGER PRIMARY KEY,
            name TEXT,
            age INTEGER,
            address TEXT
        )
""")
```

```python
cursor.execute('INSERT INTO students (name, age, address) VALUES (?, ?, ?)', ('Ram', 25, "Biratnagar"))

conn.commit()
```

```python
cursor.execute('SELECT * FROM students')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('SELECT * from students WHERE address="Itahari"')
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute('SELECT * from students WHERE name LIKE "R%"')
rows = cursor.fetchall()
for row in rows:
    print(row)
    
# R% fetches the starting with r
# %R fetches the ending with r
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite:///anotherdb.db', echo=True)
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String

Base = declarative_base()

class Student(Base):
    __tablename__ = 'students'
    id = Column(Integer, primary_key = True)
    name = Column(String)
    age = Column(Integer)
    address = Column(String)
    
#create table
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import sessionmaker
Session = sessionmaker(bind=engine)
session = Session()
```

```python
student1 = Student(name="Ram", age=25, address="Itahari")
student2 = Student(name="Shyam", age=25, address="Mahendranagar")
student3 = Student(name="Hari", age=25, address="Biratnagar")
session.add(student1)
session.add(student2)
session.add(student3)
session.commit()
```

```python
data = session.query(Student).all()
for row in data:
    print(row.id, row.name, row.age, row.address)
```

```python
data = session.query(Student).filter_by(address='Itahari')
for row in data:
    print(row.id, row.name, row.age, row.address)
```
