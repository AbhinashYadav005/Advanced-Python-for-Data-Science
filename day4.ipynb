```python
# simple inheretance: Base Class
class Animal:
    def __init__(self, name):
        self.name = name 
        
    def speak(self):
        print(f"{self.name} is Speaking")

# Derived Class
class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed
        
    def speak(self):
        super().speak()
        print(f"{self.name} of {self.breed} is barking")
        
        
        
d1 = Dog("Lucifer", "hybrid")
d1.speak()   
```

```python
# Multiple inheritance

class A:
    def show(self):
        print("A")
    
    
class B:
    def show(self):
        print("B")
    

class C(A,B): # depends on the order of class inheritance
    pass

c1 = C()
c1.show()
```

```python
# Method Resolution Order
C.mro()
```

```python
class DataList:
    def __init__(self, alist):
        self.alist = alist
        
    def mean(self):
        return sum(self.alist)/len(self.alist)
            
```

```python
l1 =  DataList([1,2,3,4,5])
l1.mean()
```

```python
# exception handling
while(True):
    try:
        n1 = int(input("Enter first number: "))
        n2 = int(input("Enter second number: "))
        result = n1/n2
        print(f"result = {result}")
    except ZeroDivisionError:
        print("Number can't be divided by zero!")
    except ValueError:
        print('Value error')
    finally:
        print('this always runs') #always run
```
