```python
import csv

data = [
    ['Name', 'Address', 'Age '],
    ['Ram', ' Biratnagar', ' 22'],
    ['Shyam', ' Itahari', ' 34'],
    ['Hari', ' Dharan', ' 27 ']
]

with open(file="newnew.csv", mode='w',newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
import csv

data = [
{'Name': 'Ram', 'Address': ' Biratnagar', 'Age ': ' 22'},
{'Name': 'Shyam', 'Address': ' Itahari', 'Age ': ' 34'},
{'Name': 'Hari', 'Address': ' Dharan', 'Age ': ' 27 '}
]

with open(file='newnewnew.csv',mode='w',newline='') as f:
    writer = csv.DictWriter(f, fieldnames=["Name","Address","Age "])
    writer.writeheader()
    writer.writerows(data)
```

```python
import pandas as pd

dataframe = pd.DataFrame([
    {"Roll": 1, "Name": "Ram", "Eng": 80,"Math":81,"Science":82},
    {"Roll": 2, "Name": "Shyam", "Eng": 79,"Math":80,"Science":83}
])

dataframe.to_csv("Marksheet.csv",index=False)
```

```python
dataframe['Total'] = dataframe['Eng']+ dataframe['Math']+dataframe['Science']
print(dataframe)
```

```python
dataframe['Percentage'] = (dataframe['Total']/3).round(decimals=3)
print(dataframe)
```
